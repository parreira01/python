### Composição dos dados

Antes de apresentar as análises de composição dos dados, é interessante observarmos a diferença entre função e atributo no Python. Ambas podem ser aplicadas sobre os objetos, mas enquanto na maioria das vezes é possível passar parâmetros que alteram o resultado final, como a função `read_csv` que resulta em diferentes dataframes caso o separador seja vírgula ou ponto e vírgula, no segundo é devolvido um atributo inerente ao objeto, sendo assim imutável.

Um atributo muito utilizado na análise de composição dos dados é o atributo **shape**.



```
# tamanho do dataframe
df_imrs.shape
```

Que devolve uma *tupla*, onde o primeiro valor corresponde ao número de linhas e o segundo valor ao número de colunas do dataframe.

```
print('Número de linhas do objeto df_imrs: {}'.format(df_imrs.shape[0]))
print('Número de colunas do objeto df_imrs: {}'.format(df_imrs.shape[1]))
```

Para dataframes, que são objetos de duas dimensões, é mais informativo utilizar o atributo shape do que a função len(). Dado que a última retorna somente o número de linhas.

```
# função len
len(df_imrs)
```

Além do shape e do len, o Python possui as seguintes funções e atributos para observar a composição dos dados.

* head(): retorna as n primeiras entradas do dataframe, onde por default n = 5. [Documentação](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.head.html).
* tail(): retorna as n últimas entradas do dataframe, onde por default n = 5. [Documentação](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.tail.html).
* sample(): retorna n entradas aleatórias do dataframe, onde por default n = 1. [Documentação](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sample.html).
* value_counts(): retorna a quantidade de cada valor único de uma **série**. [Documentação](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.value_counts.html).
* unique(): retorna os valores únicos de uma série. [Documentação](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.unique.html).
* nunique(): retorna a quantidade de valores únicos de um dataframe. [Documentação](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.nunique.html).
* dtypes: retorna o tipo de cada coluna de um dataframe. [Documentação](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.dtypes.html).
* info(): retorna um sumário dos dados. [Documentação](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.info.html).

### Filtragem dos dados

Na maioria dos casos, analisar todo um dataframe de uma vez é uma tarefa quase que impossível. Até o próprio pandas evita de mostrar todas as informações de um dataframe de uma vez, com o objetivo de não utilizar tanto a memória.

Sendo assim, é mais interessante observar dados filtrados por um valor específico para um coluna específica ou para ambos de uma só vez. Nesse caso, estamos interessados em fazer consultas ao dataframe e filtrar os dados que mais nos interessam.


```
# observando as cinco primeiras entradas do objeto
df_imrs.head()
```

```
# filtrando valores de Belo Horizonte
df_imrs['Município' == 'Belo Horizonte']
```

```
# filtrando valores de Belo Horizonte
df_imrs[df_imrs['Município'] == 'Belo Horizonte']
```

```
# filtrando valores de Belo Horizonte e Ribeirão das Neves (OR)
df_imrs[(df_imrs['Município'] == 'Belo Horizonte') | (df_imrs['Município'] == 'Ribeirão das Neves')]
```

```
# selecionando os municípios
municipios = ['Belo Horizonte', 'Ribeirão das Neves']

# utilizando a função isin
df_imrs[df_imrs['Município'].isin(municipios)]
```

```
# filtrando valores de Belo Horizonte e Ribeirão das Neves (AND)
df_imrs[(df_imrs['Município'] == 'Belo Horizonte') & (df_imrs['Município'] == 'Ribeirão das Neves')]
```

```
# filtrando valores pela qualidade da educação
df_imrs[df_imrs['Índice de Qualidade da Educação para a 4ª série /5º ano do EF (matemática)'] > 0.5]
```

```
# filtrando valores por comparação entre as variáveis
df_imrs[df_imrs['Índice de Qualidade da Educação para a 4ª série /5º ano do EF (matemática)'] >
        df_imrs['Índice de Qualidade da Educação para a 4ª série /5º ano do EF (Língua Portuguesa)']]
```

```
# filtrando somente as colunas que são importantes
colunas = ['ibge7', 'Município','Índice de Qualidade Geral da Educação','Existência de Conselho Municipal de Educação','Conselho de Educação em Atividade']

df_imrs[colunas]
```
