## Vídeo 88: Gráfico tipo Boxplot 13:09


```python
fig, ax = plt.subplots(figsize = (10,5))

enade.boxplot('nu_idade', ax = ax)
ax.set_title('Boxplot da Idade', fontsize = 14)
#ax[0].set_ylabel('Densidade')
#ax[0].set_xlabel('Idade')

fig.tight_layout
```

```python
fig, ax = plt.subplots(figsize = (10,5))

enade.boxplot('nu_idade', ax = ax, vert = False, widths = 0.6)# vert é vertical e ele como falso é horizontal.
ax.set_title('Boxplot da Idade', fontsize = 14)
ax.set_yticks([]) # a lista vazia no parentese deixa o eixo y sem informação.
ax.set_xlabel('Idade')

fig.tight_layout
```

```python
fig, ax = plt.subplots(figsize = (10,5))

enade.boxplot('nu_idade', ax = ax, vert = False, widths = 0.6, by = 'cd_catadd')# o parametro 'by' entrega os valores da coluna catadd,
# dividida na informação das instituições públicas e privadas.
# O título 'boxplot grouped by cd_catadd' é inserido automaticamente quando usado o parâmetro 'by'.

ax.set_title('Boxplot da Idade', fontsize = 14)
#ax.set_yticks([]), essa informação é retirada do código pois agora é necessário que se defina as instiuições no eixo y.
ax.set_xlabel('Idade')

fig.suptitle('') # Essa função retira o texto colocado automaticamente pelo 'by'.

fig.tight_layout
```

###Gráfico de pizza

```python
enade.head(1)
```

```python
enade['co_regiao_curso'].value_counts()
```

```python
# extrair as ocorrências para um dicionário
regioes = dict(enade['co_regiao_curso'].value_counts())

regioes
```

```python
# Para se criar um gráfico de pizza é necessário duas informações: labels / Sizes
labels = regioes.keys()
sizes = regioes.values()
```

```python
fig, ax = plt.subplots() # usa-se sempre que for fazer um gráfico

ax.pie(sizes, labels = labels, autopct='%1.2%%')# pie é a função do gráfico de pizza.
# O Parametro autopct = %1.2%% coloca a porcentagem em cada pedaço da pizza com duas casas decimais.
ax.set_title('Estudantes por Região do País')

plt.tight_layout()# usa-se sempre pra mostrar o resultado de um gráfico
```

```python
fig, ax = plt.subplots() # usa-se sempre que for fazer um gráfico

ax.pie(sizes, labels = labels, autopct='%1.2f%%')
# O Parametro autopct = %1.2f%% coloca a porcentagem em cada pedaço da pizza com duas casas decimais.
ax.set_title('Estudantes por Região do País')

plt.tight_layout()# usa-se sempre pra mostrar o resultado de um gráfico
```

```python
explode = (0.2,0.1,0.1,0,0) # a função explode afasta do centro os pedaços da pizzas, da maior para a menor porcentagem.
```

```python
fig, ax = plt.subplots()

ax.pie(sizes, labels = labels, autopct='%1.2f%%', explode = explode, shadow=True)# coloca o explode em prática
ax.set_title('Estudantes por Região do País')

plt.tight_layout()# usa-se sempre pra mostrar o resultado de um gráfico
```

##Vídeo 91: Mapa de calor (heatmap) 13:26


```python
imrs.head(1)
```

```python
variaveis = ['Índice de Qualidade da Educação para a 4ª série /5º ano do EF (matemática)',
             'Índice de Qualidade da Educação para a 8ª série/9º ano do EF (matemática)',
             'Índice de Qualidade da Educação para a 3ª série do EM (matemática)',
             'Índice de Qualidade da Educação para a 4ª série /5º ano do EF (Língua Portuguesa)',
             'Índice de Qualidade da Educação para a 8ª série/9º ano do EF (Língua Portuguesa)',
             'Índice de Qualidade da Educação para a 3ª série do EM (Língua Portuguesa)']
```

```python
imrs[variaveis]
```

```python
# Plotando o gráfico heatmap, que é um pacote específico do Seaborn
fig, ax = plt.subplots(figsize = (10,10))

sns.heatmap(round(imrs[variaveis].corr(), 2 ), ax = ax)
ax.set_title('Correlação dos índices: matemática e português')



plt.tight_layout()
```

```python
fig, ax = plt.subplots(figsize = (10,10))

sns.heatmap(round(imrs[variaveis].corr(), 2 ), ax = ax, vmax = 1, vmin = -1, cmap = 'Blues', annot = True)
# vmax e vmin mostram a escala máxima (maior correlação) e mínima (mennor correlação) da correlação;
# cmap edita as cores do gráfico, nesse caso em tons de azul;
# annot = True coloca os valores da correlação dentro do gráfico.
ax.set_title('Correlação dos índices: matemática e português')

plt.tight_layout()
```
