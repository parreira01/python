## Vídeo 66: Automatização de Tarefas - Novo Objeto Dados da PNAD


```python
def obter_pnad(caminho_dicionario, caminho_dados):
  # 1º- obter as informações do dicionario;
  dicionario = pd.read_excel(caminho_dicionario, skiprows = [0])
  larguras = list(dicionario['Tamanho'].dropna().astype(int))
  codigos = list(dicionario['Código de variável'].dropna())

  larguras.pop(1)
  codigos.pop(1)

  # 2º - importando os dados da pnad;
  # Criar um novo objeto com nome diferente dos arquivos envolvidos, para que não sejam sobescritos;
  dados_pnad =pd.read_fwf(caminho_dados, widths = larguras, header = None)

  dados_pnad.columns = codigos

  dados_pnad['UF'] = dados_pnad['V0102'].astype(str).str[0:2]

  return dados_pnad
```

## Vídeo 74: Automatização de Tarefas - IPCA: Encapsulando código


```python
# crie uma função def com o nome de ajustes_sidra.
# nova automatização é referente ao Vídeo 78: Aut. completa dos dados do SIDRA - ajustando a função de tratamento 19:59.
# Vídeo 79: Automatização completa dos dados do SIDRA - ajustando o tempo 28:30.


def ajustes_sidra(dataframe_dados):

  #criar uma cópia do dataframe para fazer ajustes sem faze-las no original.
  dados = dataframe_dados.copy()

  # 1º passo: salvar as informações do data frame com a primeira linha dass colunas.
  # depois um for com o iloc da primeira posição de linha
  # strip para eleminar todos os espaços das informações do dataframe
  nome_colunas = []
  for nome in dados.iloc[0]:
    nome_colunas.append(nome.strip())

  dados.columns = nome_colunas # adicionar nova coluna 'nome_colunas'.

  # 2º passo:
  dados.drop(0, axis = 0, inplace = True)

  # 3º passo: index temporal
  tempo = [] # aula 74
  for item in dados.columns: # nova automatização
    if item == 'Mês (Código)': # nova automatização
      for periodo in dados['Mês (Código)']: # aula 74
        texto = periodo[0:4] + '/' + periodo[-2:] # aula 74
        tempo.append(texto) # aula 74
      dados.index = pd.to_datetime(tempo)# aula 74
      dados.index = dados.index.strftime('%m-%Y')# aula 74

    elif item == 'Trimestre (Código)': # nova automatização
      for periodo in dados['Trimestre (Código)']: # nova automatização

        if periodo [-2:] == '01': # nova automatização
          per = periodo [-2:] # nova automatização
        elif periodo [-2:] == '02': # nova automatização
          per = '04' # nova automatização
        elif periodo [-2:] == '03': # nova automatização
          per = '07' # nova automatização
        elif periodo [-2:] == '04': # nova automatização
          per = '10' # nova automatização

        texto = periodo[0:4] + '/' + per # nova automatização
        tempo.append(texto) # nova automatização

      dados.index = pd.to_datetime(tempo) # Vídeoo 79
      dados.index = dados.index.to_period('Q').strftime('Q%q - %Y') # Vídeoo 79

    elif item == 'Semestre (Código)': # Vídeoo 79
      for periodo in dados ['Semestre (Código)']: # Vídeoo 79
        texto = periodo[0:4] + '/' + periodo[-2:] # Vídeoo 79
        tempo.append(texto) # Vídeoo 79

      dados.index = pd.to_datetime(tempo) # Vídeoo 79
      dados.index = dados.index.strftime('S%m-%Y') # Vídeoo 79

  # 4º passo: criar uma lista variaveis_drop
  variaveis_drop = []

  for coluna in dados.columns: # condição para retirada dascolunas com código e mês.
    if 'Código' in coluna: # Vídeoo 74
      variaveis_drop.append(coluna) # Vídeoo 74
    elif coluna == 'Mês': # Vídeoo 74
      variaveis_drop.append(coluna) # Vídeoo 74
    elif coluna == 'Trimestre': # Vídeoo 79
      variaveis_drop.append(coluna) # Vídeo 79
    elif coluna == 'Semestre': # Vídeo 79
      variaveis_drop.append(coluna) # Vídeo 79

  dados.drop(labels = variaveis_drop, axis = 1, inplace = True) # Vídeoo 74

  # 5º passo:
  dados  ['Valor'] = dados  ['Valor'].replace('-', np.nan).replace('...', np.nan).astype(float) # video 74 # o replace('-', np.nan) é do video 79

  return dados
```

```python
# vídeo 79
# O vídeo 81 foi o tratamento para receber todas as tabelas a serem importadas através do código de automação.
contador = 0
for tabela in codigos['Codigos']:
  base = 'http://api.sidra.ibge.gov.br/values'
  url = base + tabela

# Ao invés de imprimir o resultado, vamos obter os dados na API
  resposta = requests.get(url)

# Salvar os dados obtidos em um data frame
  df = pd.DataFrame(resposta.json())

# Tratamento dos dados
  df_ajustado = ajustes_sidra(df)


# Vídeo 80: Automatização completa dos dados do SIDRA - Exportando o dataframe 3
# Exportando dados
  if contador == 0:
    with pd.ExcelWriter('arquivo.xlsx') as writer:
      df_ajustado.to_excel(writer, sheet_name = 'Tabela{}'.format(contador+1))
  else:
    with pd.ExcelWriter('arquivo.xlsx', mode = 'a') as writer:
      df_ajustado.to_excel(writer, sheet_name = 'Tabela{}'.format(contador+1))

  contador += 1
```
